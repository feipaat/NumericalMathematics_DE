# <center>Interpolation</center>
## <center>Lagrange and Newton interpolation polynomials</center>

<br>
<b>Example.</b> Let us consider the data

$$(t_i, f_i):\ (−3, −209),\ (−2, −43),\ (−1, −1),\ (1, −1),\ (2, −19).$$

Fit to the data the appropriate Langrange interpolation polynomial!


```octave
t = [-3 -2 -1 1 2];
f = [-209 -43 -1 -1 -19];

% Using the built-in fuction
p = polyfit(t,f,length(t)-1);
tt = linspace(t(1),t(end),100);
ff = polyval(p,tt);

plot(t,f, 'm*',tt,ff)
title('The interpolation polynomial')
```

<br>
<b>Runge's example</b> Let us consider the function

$$f(x)=\frac{1}{1+25x^2},\ \quad x\in [-1,1].$$

Make a plot using the polynomials $L_6(x), L_{12}(x), L_{18}(x)$ on a uniform grid, the original function and the Chebyshev polynomial!

```octave
Rungeexample(-1,1,6)
```

## <center>Spline interpolation polynomial</center>



Solving the previous example using spline interpolation

```octave
 x = linspace(-1,1,100); 
 y = 1./(1+25*x.^2);
 yspline = spline (x,y,x);
 plot(x,y,'b',x,yspline,'mo');
 legend('f(x)','Spline interpolation polynomial')
```
