## <center>Bisection, False position and Newton's methods</center>

<br>
<b>Example.</b> Let us consider the equation

$$f(x)=e^x-4x^2=0,\quad x\in [3,5].$$

Determine the root of bisection method using the learned stopping criterion.

```octave
% Bisection
clear all
a=3; b=5; func='exp(x)-4*x^2'; maxit=100; TOL=10^(-6);

format long
abserror = inf;
k=0;
while k<maxit && abserror>TOL
   k=k+1;
   x=a+(b-a)/2;
      fx=eval(func);
   if fx>0 
      b=x;
   else
      a=x;
   end;
   abserror = abs(b-a)
end;
k
x
```

```octave
x = 4.2:0.01:4.4;
plot(x,exp(x)-4*x.^2)
```

```octave
format long
%[iteration_number x_numerical_value] = falseposition(func,a,b,maxit,TOL)
[iteration_number x_numerical_value] = falseposition('exp(x)-4*x^2',3,5,100,1e-6)
```

```octave
% [iteration_number x_numerical_value] = newtonmethod(func,derivative,a,b,x,maxit,TOL)
[iteration_number x_numerical_value] = newtonmethod('exp(x)-4*x^2','exp(x)-8*x',3,5,4,100,1e-6)
```
