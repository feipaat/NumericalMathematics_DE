## <center>Local extrama of two-variable functions</center>

<br>
<b>Example 1.</b> Let us consider the two-variable function

$$f(x_1,x_2)=\frac{1}{4}\left(2x_1^3+3x_1^2-12x_1\right)+\frac{1}{2}\left(2x_2^3+3x_2^2-12x_2\right).$$

Analyze it in terms of contour lines and gradients.


```octave
x_1 = -3:0.05:2;
x_2 = -3:0.05:2;

f=@(x_1,x_2)((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[X_1, X_2] = meshgrid(x_1, x_2);
surf(X_1, X_2, f(X_1,X_2));
figure;
contourf(X_1, X_2, f(X_1,X_2));
```

```octave
[C,h] = contour(X_1,X_2,f(X_1,X_2));
clabel(C,h,'FontSize',12,'Color','black')
```

```octave
figure
[x_1,x_2] = meshgrid(-3:.2:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(Z);
[C,h] = contour(x_1,x_2,Z);
clabel(C,h,'FontSize',12,'Color','black')
hold on
[x_1,x_2] = meshgrid(-3:.2:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(Z);
quiver(x_1,x_2,DX,DY)
title('The contour lines and the positive gradient field')
hold off
```

```octave
figure
[x_1,x_2] = meshgrid(-3:.2:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(-Z);
[C,h] = contour(x_1,x_2,Z);
clabel(C,h,'FontSize',12,'Color','black')
hold on
[x_1,x_2] = meshgrid(-3:.6:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(-Z);
quiver(x_1,x_2,DX,DY)
title('The contour lines and the negative gradient field')
hold off
```

<b>Example.</b> Let us consider the two-variable function $f(x_1,x_2)=x_1^2-x_2^2$ .

```octave
x_1 = -2:0.1:2;
x_2 = -2:0.1:2;
f=@(x_1,x_2)(x_1.^2-x_2.^2);
[X_1, X_2] = meshgrid(x_1, x_2);
surf(X_1, X_2, f(X_1,X_2));
```

```octave
x_1 = -3:0.05:2;
x_2 = -3:0.05:2;
x=@(x_1,x_2)((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
colormap("white")
[X_1, X_2] = meshgrid(x_1, x_2);
hold on
contourf(X_1, X_2, x(X_1,X_2));
plot(1,1,'m*',1,-2,'m*',-2,1,'m*',-2,-2,'m*')
hold off
```

```octave
x_1 = -3:0.05:2;
x_2 = -3:0.05:2;
x=@(x_1,x_2)((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
colormap("white")
[X_1, X_2] = meshgrid(x_1, x_2);
hold on
contourf(X_1, X_2, x(X_1,X_2));
plot(1,1,'m*',1,-2,'m*',-2,1,'m*',-2,-2,'m*')
[x_1,x_2] = meshgrid(-3:.4:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(Z);
quiver(x_1,x_2,DX,DY)
title('The contour lines and the positive gradient field','Color','magenta')
[x_1,x_2] = meshgrid(-3:.2:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(Z);
[C,h] = contour(x_1,x_2,Z);
clabel(C,h,'FontSize',12,'Color','magenta')
hold off

figure;
x_1 = -3:0.05:2;
x_2 = -3:0.05:2;
x=@(x_1,x_2)((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[X_1, X_2] = meshgrid(x_1, x_2);
surf(X_1, X_2, x(X_1,X_2));
```
